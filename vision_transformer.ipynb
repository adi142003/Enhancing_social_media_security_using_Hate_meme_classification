```python
from PIL import Image
import os
small_image_count = 0
small_width_count = 0
small_height_count = 0
folder_path = './DATA/testImages'
max_width=250
max_height=250


# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    # Construct full file path
    file_path = os.path.join(folder_path, filename)
    
    # Check if the file is an image
    try:
        with Image.open(file_path) as img:
            width, height = img.size
            
            # Check if the image dimensions are smaller than the specified size
            if width <= max_width and height <= max_height:
                small_image_count += 1
            elif width <= max_width and height >=max_height:
                small_width_count += 1
            elif width >= max_width and height <= max_height:
                small_height_count += 1
    except IOError:
        # If the file is not an image, skip it
        print(f"Skipping non-image file: {filename}")
        continue

print(f"Number of images smaller than 224x224: {small_image_count}")
print(f"Number of images width smaller than 224x224: {small_width_count}")
print(f"Number of images height smaller than 224x224: {small_height_count}")

```

```python
import matplotlib.pyplot as plt
widths = []
heights = []

# Iterate over all files in the folder
for filename in os.listdir(folder_path):
    # Construct full file path
    file_path = os.path.join(folder_path, filename)
    
    # Check if the file is an image
    try:
        with Image.open(file_path) as img:
            width, height = img.size
            widths.append(width)
            heights.append(height)
    except IOError:
        # If the file is not an image, skip it
        print(f"Skipping non-image file: {filename}")
        continue
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(widths, heights,alpha=0.5)
plt.title('image dimensions')
plt.xlabel('width')
plt.ylabel('height')
plt.grid(True)
plt.show()
```
